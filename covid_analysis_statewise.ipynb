```python


df = pd.read_csv("/content/covid_state_level_latest.csv").head(5)
fig_1 = px.scatter(df, x="State", y="Confirmed", color="State")
fig_1.show()
```

```python
fig_2 = px.scatter(df, x="State", y="Deaths", color="State")
fig_2.show()
```

```python
fig_3=px.parallel_coordinates(df,labels={"Deaths":"DEATHS_CASES"},color="Confirmed")
fig_3.show()
```

```python
fig_4=px.scatter(df,x="State",y="Confirmed",color="State",marginal_y="violin",marginal_x="box",trendline="ols",template="simple_white")
fig_4.show()
```

```python
fig_5=px.scatter_3d(df,x="Confirmed",y="Active",z="Recovered",color="State",size="Confirmed",symbol="State",opacity=1)
fig_5.show()
```

```python
fig_6=px.scatter_matrix(df,dimensions=["Deaths", "Confirmed", "Active", "Recovered"],color="State")
fig_6.show()
```

```python
fig_7=px.bar(df,x="State",y="Confirmed",color="State")
fig_7.show()
```

```python
fig_8=px.histogram(df,x="Confirmed",color="State")
fig_8.show()
```

```python
fig_9=px.funnel(df,x="Confirmed",y="State")
fig_9.show()
```

```python
fig_10=px.sunburst(df,path=["Deaths","Active","Recovered"],values="Confirmed",color="State")
fig_10.show()
```
